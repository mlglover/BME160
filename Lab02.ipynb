### seqCleaner

```python
#!/usr/bin/env python3 
# Name: Marissa Glover (mlglover) 
# Group Members: Parissa Monem (pmonem) and Matt Modena (mmodena)

'''
Read a DNA string from user input and return a collapsed substring of embedded Ns to: {count}.

Example: 
 input: AaNNNNNNGTC
output: AA{6}GTC

Any lower case letters are converted to uppercase.
'''

class DNAstring (str):
    '''Determine the number of Ns in a string and create a new object.
    A single run of Ns will be collapsed and show {#Ns}.'''
    def purify(self):
        '''Given an upper case string, this will find the first and last N.
        The string will be split at the Ns and then merged back together with {#Ns} in place of the Ns.'''
        numN = self.count('N')
        numNstring = str({numN})
        findFirstN = self.find('N')
        findLastN = self.rfind('N')
        splitSeq = self.split('N')
        beforeN = splitSeq[0]
        afterN = splitSeq[-1]
        pureDNA = ''.join([beforeN, numNstring, afterN])
        return pureDNA
    
def main():
    '''Get user DNA data and clean it up. Make it upper case before passing to the class.'''
    data = input('DNA data? ')
    upperDNA = data.upper()
    thisDNA = DNAstring (upperDNA)
    pureData = thisDNA.purify()
    print (pureData)
    
main()
```

In this exercise, you will create a program that takes three sets of atomic coordinates, all provided on a **single line**. The program then calculates the bond lengths and angles. For this program, you can start with the Triad class (provided). Your task is to create a Python program called coordinateMathSoln that:

 - outputs the N-C and N-Ca bond lengths and the C-N-Ca bond angle with correct number of significant digits (see below)
 
For example, if I enter the following coordinates (notice.. they are all on one line !!!) :<br>

C = (39.447, 94.657, 11.824) N = (39.292, 95.716, 11.027) Ca = (39.462, 97.101, 11.465) <br>

(Note: make sure that the angle returned is in **degrees** !!)
### Hints: 
 - Each coordinate will contain only 3 numeric values. 
 - Bond lengths are the distance between two points, which for points P and Q in 3-space,
$(P_x , P_y , P_z )$ and $(Q_x , Q_y , Q_z )$ respectively, the distance between them is:


$$
\begin {eqnarray}
     \|PQ\| &=&\sqrt{(P_x-Q_x)^2 + (P_y-Q_y)^2 + (P_z-Q_z)^2)} \\
            &=&\sqrt{\sum^{i \in x,y,z}{(P_i-Q_i)^2}} \\
\end{eqnarray}
$$

 - Bond angles can be calculated from the dot product.
 
Let’s say that we have three points in space labeled P, Q and R. We are interested in the angle at point Q that is made from line segments QP and QR.
The dot product tells us (for standard vectors P and R) that:
$$P \cdot R=\|P\| \|R\| \cos{\theta}$$
in this notation,$\|P\|$ refers to the length of vector P as a standard vector (assumed to begin at the origin (0,0,0) ).
We can then see that the angle between vectors P and R can be found by:
$$\cos{\theta}=\frac{P \cdot R} {\|P\| \|R\|}$$

We can calculate the dot product using the sum of products of the vector components:
$$P \cdot R = \sum^{i \in x,y,z}{P_i R_i} $$

Now, to find vector P in standard form, we need to remember that QP starts at Q, so we need to place the origin at Q and find out where P is in that new space. We do that by subtracting the components of Q from P.
Putting all of this together, we get:
$$\theta=\cos^{-1}{\frac{\sum^{i \in x,y,z}{(P_i - Q_i) (R_i - Q_i) }} {\|QP\| \|QR\|}}$$
Remember, $\theta$ is in radians.

Below I have given you a class (Triad) with methods to calculate dot products (dot), dot products of translated vectors (ndot), distances (dPQ, dPR, dPR) and angles in radians (angleP, angleQ and angleR) for each of the three points in a Triad object.

### coordinateMathSoln

```python
#!/usr/bin/env python3 
# Name: Marissa Glover (mlglover)
# Group Members: Parissa Monem (pmonem) and Matt Modena (mmodena)

'''
Calculate distances between points and angles given a triad of protein coordinates.

input: x,y,z coordinates for a carbon, nitrogen, and alpha carbon of a protein. 
    (example: C = (39.447, 94.657, 11.824) N = (39.292, 95.716, 11.027) Ca = (39.462, 97.101, 11.465))
output: N-C bond length, N-Ca bond length, and C-N-Ca bond angle in degrees.
    (example:   N-C bond length = 1.33
                N-Ca bond length = 1.46
                C-N-Ca bond angle = 124.0)
'''

import math
class Triad :
    """
    Calculate angles and distances among a triad of points.
 
    Author: David Bernick
    Date: March 21, 2013
    Points can be supplied in any dimensional space as long as they are consistent.
    Points are supplied as tupels in n-dimensions, and there should be three
    of those to make the triad. Each point is positionally named as p,q,r
    and the corresponding angles are then angleP, angleQ and angleR.
    Distances are given by dPQ(), dPR() and dQR()
 
    Required Modules: math
    initialized: 3 positional tuples representing Points in n-space
             p1 = Triad( p=(1,0,0), q=(0,0,0), r=(0,1,0) )
    attributes: p,q,r the 3 tuples representing points in N-space
    methods:  angleP(), angleR(), angleQ() angles measured in radians
          dPQ(), dPR(), dQR() distances in the same units of p,q,r
 
    """
 
    def __init__(self,p,q,r) :
        """ Construct a Triad. 
        
        Example construction:
            p1 = Triad( p=(1.,0.,0.), q=(0.,0.,0.), r=(0.,0.,0.) ). 
        """
        self.p = p
        self.q = q
        self.r = r
# private helper methods
    def d2 (self,a,b) : # calculate squared distance of point a to b
        return float(sum((ia-ib)*(ia-ib)  for  ia,ib in zip (a,b)))
    
    def dot (self,a,b) : # dotProd of standard vectors a,b
        return float(sum(ia*ib for ia,ib in zip(a,b)))
    
    def ndot (self,a,b,c) : # dotProd of vec. a,c standardized to b
        return float(sum((ia-ib)*(ic-ib) for ia,ib,ic in zip (a,b,c)))
    
# calculate lengths(distances) of segments PQ, PR and QR
    def dPQ (self):
        """ Provides the distance between point p and point q """
        return math.sqrt(self.d2(self.p,self.q))
    
    def dPR (self):
        """ Provides the distance between point p and point r """
        return math.sqrt(self.d2(self.p,self.r))
    
    def dQR (self):
        """ Provides the distance between point q and point r """
        return math.sqrt(self.d2(self.q,self.r))
    
    def angleP (self) :
        """ Provides the angle made at point p by segments pq and pr (radians). """
        return math.acos(self.ndot(self.q,self.p,self.r) /   math.sqrt(self.d2(self.q,self.p)*self.d2(self.r,self.p)))
    
    def angleQ (self) :
        """ Provides the angle made at point q by segments qp and qr (radians). """
        return math.acos(self.ndot(self.p,self.q,self.r) /  math.sqrt(self.d2(self.p,self.q)*self.d2(self.r,self.q)))
 
    def angleR (self) :
        """ Provides the angle made at point r by segments rp and rq (radians). """
        return math.acos(self.ndot(self.p,self.r,self.q) /  math.sqrt(self.d2(self.p,self.r)*self.d2(self.q,self.r)))

def main():
    '''Ask for an input with three sets of coordinates and parse the string.
    Replace parentheses and equals signs with commas, and replace spaces with nothing.
    Then we can split the line at the commas to get a list of strings.'''
    dataInput = input('Three sets of coordinates: ')
    splitLine = dataInput.replace('(',',').replace(')',',').replace('=',',').replace(' ','').split(',')
    #Use map function to turn list of strings into list of floats
    #https://www.thecodingforums.com/threads/how-to-convert-a-list-of-strings-to-a-tuple-of-floats.684531/
    p = tuple(map(float, splitLine[2:5]))
    q = tuple(map(float, splitLine[7:10]))
    r = tuple(map(float, splitLine[12:15]))
    triad = Triad(p,q,r)
    print ('N-C bond length = {0:0.2f}'.format(triad.dPQ()))
    print ('N-Ca bond length = {0:0.2f}'.format(triad.dQR()))
    #Convert C-N-Ca bond angle from radians to degrees with math
    print ('C-N-Ca bond angle = {0:0.1f}'.format(math.degrees(triad.angleQ())))

main()
```

### Triad rewrite

```python
#!/usr/bin/env python3 
# Name: Marissa Glover (mlglover)
# Group Members: None
'''
Create a different version of triad that can be used as an alternative to the previous Triad class.


input: x,y,z coordinates for a carbon, nitrogen, and alpha carbon of a protein. 
    (example: C = (39.447, 94.657, 11.824) N = (39.292, 95.716, 11.027) Ca = (39.462, 97.101, 11.465))
output: N-C bond length, N-Ca bond length, and C-N-Ca bond angle in degrees.
    (example:   N-C bond length = 1.33
                N-Ca bond length = 1.46
                C-N-Ca bond angle = 124.0)
'''

#import sys
#!conda install --yes --prefix {sys.prefix} numpy
import numpy as np
import math
class Triad:
    def __init__(self,p,q,r) :
        """ Construct a Triad.

        Convert the tuple for coordinates of each atom into array.
        Needs to use asarray function, not array (array will not convert input will only create new array).
        
        Example construction:
            p1 = Triad( p=(1.,0.,0.), q=(0.,0.,0.), r=(0.,0.,0.) ).
        """
        self.p = np.asarray(p)
        self.q = np.asarray(q)
        self.r = np.asarray(r)
        
    def dPQ (self):
        """ Provides the distance between point p and point q.
        Find the difference between p and q points in space which have been converted to arrays (2D to 3D).
        
        Use the linear algebra functions from NumPy to use standard linear algebra algorithms.
        (from https://stackoverflow.com/questions/1401712/how-can-the-euclidean-distance-be-calculated-with-numpy)
        
        A norm is a function that satisfies certain properties pertaining to scalability and additivity,
        and assigns a strictly positive real number to each vector in a vector space over the field of real or complex numbers
        —except for the zero vector, which is assigned zero (from https://en.wikipedia.org/wiki/Norm_(mathematics)).
        Analogous to absolute value.
        linalg.norm(x[, ord, axis, keepdims]) for matrix or vector norm.
        """
        return np.linalg.norm(self.p - self.q)
    
    def dPR (self):
        """ Provides the distance between point p and point r.
        The same as the dPQ function, but for points p and r.
        """
        return np.linalg.norm(self.p - self.r)
    
    def dQR (self):
        """ Provides the distance between point q and point r.
        The same as the dPQ function, but for points q and r.
        """
        return np.linalg.norm(self.q - self.r)
    
    def angleP (self) :
        """ Provides the angle made at point p by segments pq and pr (radians).
        
        cosAngle = (PQ.PR)/(|PQ||PR|)
        Use dot function to get dot product between QP and QR (input as arrays).
            This must be achieved by doing dot products of point q - point p and point r - point p.
            The point for the angle we are looking for needs to come last in subtraction.
            point q - point p is not the same as point p - point q.
            (Subtraction of vectors from https://www.python-course.eu/matrix_arithmetic.php)
        Then divide by the product of PQ and PR.
        To solve for the angle, the function must determine the inverse cosine.
        """
        distPQ = np.linalg.norm(self.p - self.q)
        distPR = np.linalg.norm(self.p - self.r)
        q_minus_p = self.q - self.p
        r_minus_p = self.r - self.p
        return math.acos(np.dot(q_minus_p, r_minus_p) / (distPQ * distPR))
    
    def angleQ (self) :
        """ Provides the angle made at point q by segments qp and qr (radians).
        The same as the dPQ function, but for points p and r.
        """
        distPQ = np.linalg.norm(self.p - self.q)
        distQR = np.linalg.norm(self.q - self.r)
        p_minus_q = self.p - self. q
        r_minus_q = self.r - self.q
        return math.acos(np.dot(p_minus_q, r_minus_q) / (distPQ * distQR))
        
    def angleR (self) :
        """ Provides the angle made at point r by segments rp and rq (radians).
        The same as the dPQ function, but for points p and r."""
        distPR = np.linalg.norm(self.p - self.r)
        distQR = np.linalg.norm(self.q - self.r)
        p_minus_r = self.p - self.r
        q_minus_r = self.q - self.r
        return math.acos(np.dot(p_minus_r, q_minus_r) / (distPR * distQR))

def main():
    '''Ask for an input with three sets of coordinates and parse the string.
    Replace parentheses and equals signs with commas, and replace spaces with nothing.
    Then we can split the line at the commas to get a list of strings.'''
    dataInput = input('Three sets of coordinates: ')
    splitLine = dataInput.replace('(',',').replace(')',',').replace('=',',').replace(' ','').split(',')
    #Use map function to turn list of strings into list of floats
    #https://www.thecodingforums.com/threads/how-to-convert-a-list-of-strings-to-a-tuple-of-floats.684531/
    p = tuple(map(float, splitLine[2:5]))
    q = tuple(map(float, splitLine[7:10]))
    r = tuple(map(float, splitLine[12:15]))
    triad = Triad(p,q,r)
    print ('N-C bond length = {0:0.2f}'.format(triad.dPQ()))
    print ('N-Ca bond length = {0:0.2f}'.format(triad.dQR()))
    #Convert C-N-Ca bond angle from radians to degrees with math
    print ('C-N-Ca bond angle = {0:0.1f}'.format(math.degrees(triad.angleQ())))

main()     
```

```python
import math
class Triad :
    """
    Calculate angles and distances among a triad of points.
 
    Author: David Bernick
    Date: March 21, 2013
    Points can be supplied in any dimensional space as long as they are consistent.
    Points are supplied as tupels in n-dimensions, and there should be three
    of those to make the triad. Each point is positionally named as p,q,r
    and the corresponding angles are then angleP, angleQ and angleR.
    Distances are given by dPQ(), dPR() and dQR()
 
    Required Modules: math
    initialized: 3 positional tuples representing Points in n-space
             p1 = Triad( p=(1,0,0), q=(0,0,0), r=(0,1,0) )
    attributes: p,q,r the 3 tuples representing points in N-space
    methods:  angleP(), angleR(), angleQ() angles measured in radians
          dPQ(), dPR(), dQR() distances in the same units of p,q,r
 
    """
 
    def __init__(self,p,q,r) :
        """ Construct a Triad. 
        
        Example construction:
            p1 = Triad( p=(1.,0.,0.), q=(0.,0.,0.), r=(0.,0.,0.) ). 
        """
        self.p = p
        self.q = q
        self.r = r
# private helper methods
    def d2 (self,a,b) : # calculate squared distance of point a to b
        return float(sum((ia-ib)*(ia-ib)  for  ia,ib in zip (a,b)))
    
    def dot (self,a,b) : # dotProd of standard vectors a,b
        return float(sum(ia*ib for ia,ib in zip(a,b)))
    
    def ndot (self,a,b,c) : # dotProd of vec. a,c standardized to b
        return float(sum((ia-ib)*(ic-ib) for ia,ib,ic in zip (a,b,c)))
    
# calculate lengths(distances) of segments PQ, PR and QR
    def dPQ (self):
        """ Provides the distance between point p and point q """
        return math.sqrt(self.d2(self.p,self.q))
    
    def dPR (self):
        """ Provides the distance between point p and point r """
        return math.sqrt(self.d2(self.p,self.r))
    
    def dQR (self):
        """ Provides the distance between point q and point r """
        return math.sqrt(self.d2(self.q,self.r))
    
    def angleP (self) :
        """ Provides the angle made at point p by segments pq and pr (radians). """
        return math.acos(self.ndot(self.q,self.p,self.r) /   math.sqrt(self.d2(self.q,self.p)*self.d2(self.r,self.p)))
    
    def angleQ (self) :
        """ Provides the angle made at point q by segments qp and qr (radians). """
        return math.acos(self.ndot(self.p,self.q,self.r) /  math.sqrt(self.d2(self.p,self.q)*self.d2(self.r,self.q)))
 
    def angleR (self) :
        """ Provides the angle made at point r by segments rp and rq (radians). """
        return math.acos(self.ndot(self.p,self.r,self.q) /  math.sqrt(self.d2(self.p,self.r)*self.d2(self.q,self.r)))

def main():
    '''Ask for an input with three sets of coordinates and parse the string.
    Replace parentheses and equals signs with commas, and replace spaces with nothing.
    Then we can split the line at the commas to get a list of strings.'''
    dataInput = input('Three sets of coordinates: ')
    splitLine = dataInput.replace('(',',').replace(')',',').replace('=',',').replace(' ','').split(',')
    p = tuple(map(float, splitLine[2:4]))
    q = tuple(map(float, splitLine[6:8]))
    r = tuple(map(float, splitLine[10:12]))
    print (p,q,r)
    triad = Triad(p,q,r)
    print ('felton to santa cruz = {0:0.2f}'.format(triad.dPQ()))
    print ('felton to campbell = {0:0.2f}'.format(triad.dPR()))
    #print ('C-N-Ca bond angle = {0:0.1f}'.format(math.degrees(triad.angleQ())))
    #felton = (37.0513, -122.0733) santacruz = (36.984532, -121.963123) campbell = (37.2871651, -121.9499568)
main()
```
